```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
#import warnings
#warnings.filterwarnings("ignore")


def plot_timeseries(df, time_col, unit_col, metric_col, treatment_unit, treatment_timestamp=None, 
                    figsize=(8, 4), tick_rotation=None):
    """
    Plots the timeseries for each unit.

    Parameters:
    - df: pandas DataFrame containing the data.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - metric_col: str, the name of the metric column to be analyzed.
    - treatment_unit: str, the unit that received the treatment.
    - treatment_timestamp: Timestamp, the first timestamp of the treatment period.
    - figsize: tuple, the size of the figure.
    - tick_rotation: angle of rotation of timestamp labels
    """
    # Ensure the time column is in datetime format
    df[time_col] = pd.to_datetime(df[time_col])

    # Create the plot using plt.subplots
    fig, ax = plt.subplots(figsize=figsize)

    # Get unique units
    units = df[unit_col].unique()
    # Ensure treatment_unit is first in the list if it is not NaN
    if treatment_unit in units:
        units = [treatment_unit] + [unit for unit in units if unit != treatment_unit]

    # Plot each unit
    for unit in units:
        unit_data = df[df[unit_col] == unit]
        if unit == treatment_unit:
            ax.plot(unit_data[time_col], unit_data[metric_col], label=unit, linewidth=2.5)
        else:
            ax.plot(unit_data[time_col], unit_data[metric_col], label=unit, alpha=0.9)

    # Optionally highlight the treatment period
    if treatment_timestamp:
        # Convert treatment_timestamp to pd.Timestamp if not already
        treatment_timestamp = pd.Timestamp(treatment_timestamp)
        ax.axvline(x=treatment_timestamp, color='black', linestyle='--', label='Treatment Start')

    ax.set(xlabel=time_col, ylabel=metric_col)
    if tick_rotation: ax.tick_params(axis='x', rotation=45)
    ax.legend()
    ax.grid(True, alpha=0.5, linestyle='--')

    plt.show()


def convert_bool_to_int(df):
    """
    Convert all boolean columns in a DataFrame to integers.

    Parameters:
    - df: pandas DataFrame containing the data.

    Returns:
    - df: pandas DataFrame with boolean columns converted to integers.
    """
    # Identify boolean columns
    bool_cols = df.select_dtypes(include=[bool]).columns

    # Convert boolean columns to integers
    df[bool_cols] = df[bool_cols].astype(int)
    #df.loc[:, bool_cols] = df.loc[:, bool_cols].astype(int)

    return df


def prepare_fixed_effects_data(df, time_col, unit_col, metric_col, treatment_timestamp, treatment_unit):
    """
    Prepare the data for a fixed effects regression using statsmodels.

    Parameters:
    - df: pandas DataFrame containing the data.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - metric_col: str, the name of the metric column to be analyzed.
    - treatment_timestamp: Timestamp, the timestamp indicating the treatment period.
    - treatment_unit: str, the unit that received the treatment.

    Returns:
    - prepared_df: pandas DataFrame prepared for fixed effects regression.
    """
    # Ensure the time column is in datetime format
    df[time_col] = pd.to_datetime(df[time_col])

    # Create the treatment variable
    df['After'] = df[time_col] >= treatment_timestamp
    df['Treated'] = (df[unit_col] == treatment_unit) & df['After']

    # Preserve the original unit column for clustering before creating dummies
    df['original_unit'] = df[unit_col]

    # Create dummy variables for entity and time fixed effects
    df = pd.get_dummies(df, columns=[unit_col, time_col], drop_first=True)

    # Handle any column renaming conflicts due to dummies
    df.columns = df.columns.str.replace('-', '_').str.replace(':', '_').str.replace(' ', '_')

    # Select relevant columns for the regression
    fixed_effects_cols = [col for col in df.columns if col.startswith('unit_') or col.startswith(time_col + '_')]
    prepared_df = df[[metric_col, 'Treated'] + fixed_effects_cols + ['original_unit']]

    prepared_df = convert_bool_to_int(prepared_df)
    return prepared_df


def fit_fixed_effects_model(df_fe, metric_col, time_col, clustered_errors=False, model_summary=False):
    """
    Fit a fixed effects regression model using statsmodels and output the DID coefficient, p-value, and confidence interval.

    Parameters:
    - df_fe: pandas DataFrame prepared for fixed effects regression.
    - metric_col: str, the name of the metric column to be analyzed.
    - clustered_errors: boolean, whether to use clustered standard errors.
    - model_summary: boolean, whether to print the model summary.

    Returns:
    - did_coefficient: float, the coefficient of the DID estimate.
    - p_value: float, the p-value of the DID estimate.
    - confidence_interval: tuple, the confidence interval of the DID estimate.
    """
    # Construct the formula for the regression model
    fixed_effects_terms = ' + '.join([col for col in df_fe.columns if col.startswith('unit_') or col.startswith(time_col+'_')])
    formula = f'{metric_col} ~ Treated + {fixed_effects_terms}'
    #print("Formula for regression:", formula)  # debugging
    # Fit the model using statsmodels
    if clustered_errors:
        model = smf.ols(formula, data=df_fe).fit(cov_type='cluster', cov_kwds={'groups': df_fe['original_unit']})
    else:
        model = smf.ols(formula, data=df_fe).fit()

    # Print the model summary if requested
    if model_summary:
        print(model.summary())

    # Extract the DID coefficient, p-value, and confidence interval
    did_coefficient = model.params['Treated']
    p_value = model.pvalues['Treated']
    confidence_interval = model.conf_int().loc['Treated'].values

    return did_coefficient, p_value, confidence_interval, model


def bootstrap_sample(df, metric_col, time_col, unit_col, n_bootstrap, time_initial):
    """
    Bootstrap samples consecutive timestamps from the original data and aligns them with a new increasing timestamp.

    Parameters:
    - df: pandas DataFrame, the panel data.
    - metric_col: str, the name of the metric column to analyze.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - n_bootstrap: int, the number of consecutive timestamps to bootstrap.
    - time_initial: pd.Timestamp, the starting timestamp for the bootstrapped data.

    Returns:
    - bootstrapped_df: pandas DataFrame with a continuous time series and all units for each timestamp.
    """
    # Sort the original data by time to ensure continuity
    df = df.sort_values(time_col)

    # Infer the timestamp cadence
    original_timestamps = df[time_col].unique()
    cadence = original_timestamps[1] - original_timestamps[0]  # Assumes uniform spacing

    # Bootstrap timestamps
    bootstrapped_timestamps = np.random.choice(original_timestamps, size=n_bootstrap, replace=True)

    # Initialize an empty DataFrame for the bootstrapped data
    bootstrapped_df = pd.DataFrame(columns=df.columns)

    # Create a new continuous time series starting from `time_initial`
    new_timestamps = [time_initial + i * cadence for i in range(n_bootstrap)]

    # Loop over the bootstrapped timestamps
    for i, timestamp in enumerate(bootstrapped_timestamps):
        # Take the slice of df for the current timestamp
        slice_df = df[df[time_col] == timestamp].copy()

        # Create a fake date starting from time_initial and adding a cadence every step
        slice_df[time_col] = new_timestamps[i]

        # Append the slice to bootstrapped_df
        bootstrapped_df = pd.concat([bootstrapped_df, slice_df], ignore_index=True)
    return bootstrapped_df


def create_data(df, metric_col, time_col, unit_col, treatment_unit, period_length, mde_pc):
    """
    Creates synthetic panel data from bootstrapped samples with pre- and post-periods.

    Parameters:
    - df: pandas DataFrame, the original panel data.
    - metric_col: str, the name of the metric column to analyze.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - treatment_unit: str, the unit that receives the treatment.
    - period_length: int, the number of days in the pre- and post-treatment periods.
    - mde_pc: float, the minimum detectable effect as a percentage of the treatment's pre-period average.

    Returns:
    - synthetic_data: pandas DataFrame containing the synthetic panel data.
    - treatment_timestamp: pd.Timestamp, the timestamp where the treatment starts.
    """
    time_initial = df[time_col].min()
    # Generate bootstrap samples
    synthetic_data = bootstrap_sample(df, metric_col, time_col, unit_col, n_bootstrap=2*period_length,
                                       time_initial=time_initial)
    
    # Define pre- and post-treatment periods
    pre_period = synthetic_data[time_col].unique()[:period_length]
    post_period = synthetic_data[time_col].unique()[period_length:]

    # Determine the treatment start timestamp
    treatment_timestamp = post_period[0]

    # Apply the minimum detectable effect to the treatment unit in the post-period
    pre_treatment_avg = synthetic_data[(synthetic_data[unit_col] == treatment_unit) &
                                       (synthetic_data[time_col].isin(pre_period))][metric_col].mean()
    mde = pre_treatment_avg * mde_pc

    synthetic_data.loc[(synthetic_data[unit_col] == treatment_unit) &
                       (synthetic_data[time_col].isin(post_period)), metric_col] += mde

    return synthetic_data, treatment_timestamp


def est_model(df, metric_col, time_col, unit_col, treatment_timestamp, treatment_unit, clustered_errors=True):
    """
    Estimates a fixed effects model on the synthetic data.

    Parameters:
    - df: pandas DataFrame, the synthetic panel data.
    - metric_col: str, the name of the metric column.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - treatment_timestamp: Timestamp, the start of the treatment period.
    - treatment_unit: str, the unit that received the treatment.
    - clustered_errors: bool, default is True

    Returns:
    - p_value: float, the p-value of the treatment effect.
    """
    prepared_df = prepare_fixed_effects_data(df, time_col, unit_col, metric_col, treatment_timestamp, treatment_unit)
    _, p_value, _, _ = fit_fixed_effects_model(prepared_df, metric_col, time_col, clustered_errors)
    return p_value


def simulate_power(df, metric_col, time_col, unit_col, treatment_unit,
                   period_length, mde_pc, n_iterations, α, clustered_errors=True):
    """
    Simulates power by repeating model estimation and calculating the proportion of significant results.

    Parameters:
    - df: pandas DataFrame, the original panel data.
    - metric_col: str, the name of the metric column.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - treatment_unit: str, the unit that receives the treatment.
    - period_length: int, the number of days in the pre- and post-treatment periods.
    - mde_pc: float, the minimum detectable effect as a percentage of the treatment's pre-period average.
    - n_iterations: int, the number of simulation iterations.
    - α: float, the significance level.
    - clustered_errors: bool, default is True

    Returns:
    - power: float, the estimated power of the test.
    """
    significant_count = 0

    for i in range(n_iterations):
        # Create synthetic data
        synthetic_data, treatment_timestamp = create_data(
            df, metric_col, time_col, unit_col, treatment_unit, period_length, mde_pc)

        # Estimate the model and get the p-value
        p_value = est_model(synthetic_data, metric_col, time_col, unit_col, treatment_timestamp, treatment_unit)

        # Check if the p-value is significant
        if p_value < α:
            significant_count += 1

    # Calculate power
    power = significant_count / n_iterations
    return power
```

```python
import numpy as np
import pandas as pd

def generate_trend_dataset(start_date_pre, end_date_pre, start_date_post, 
                           end_date_post, treatment_unit, control_units,
                           metric_slope, treatment_shift, random_noise_scale, shifts):
    """
    Generate a synthetic dataset for unit testing.

    Parameters:
    - start_date_pre: str, the start date of the pre-treatment period.
    - end_date_pre: str, the end date of the pre-treatment period.
    - start_date_post: str, the start date of the post-treatment period.
    - end_date_post: str, the end date of the post-treatment period.
    - treatment_unit: str, the name of the treatment unit.
    - control_units: list of str, the names of the control units.
    - metric_slope: float, the linear trend slope.
    - treatment_shift: float, the rigid shift post-treatment.
    - random_noise_scale: float, the scale of the random noise.
    - shifts: list of float, the shifts to apply to each unit.

    Returns:
    - pd.DataFrame, the synthetic dataset.
    """
    # Generate pre-treatment and post-treatment timestamps
    pre_period = pd.date_range(start=start_date_pre, end=end_date_pre)
    post_period = pd.date_range(start=start_date_post, end=end_date_post)

    def generate_unit_data(unit_name, is_treated=False):
        """
        Generate synthetic data for a single unit.

        Parameters:
        - unit_name: str, the name of the unit.
        - is_treated: bool, whether the unit is treated.

        Returns:
        - pd.DataFrame, the synthetic data for the unit.
        """
        # Pre-treatment period
        pre_metrics = metric_slope * np.arange(len(pre_period))
        # Post-treatment period
        if is_treated:
            post_metrics = metric_slope * np.arange(len(pre_period), len(pre_period) + len(post_period)) + treatment_shift
        else:
            post_metrics = metric_slope * np.arange(len(pre_period), len(pre_period) + len(post_period))

        # Combine pre- and post-periods
        metrics = np.concatenate([pre_metrics, post_metrics])
        timestamps = np.concatenate([pre_period, post_period])

        # Create DataFrame
        return pd.DataFrame({
            "timestamp": timestamps,
            "unit": unit_name,
            "metric": metrics
        })

    # Generate data for the treatment unit
    treated_data = generate_unit_data(treatment_unit, is_treated=True)

    # Generate data for the control units
    control_data = pd.concat([generate_unit_data(unit) for unit in control_units])

    # Combine all data
    df = pd.concat([treated_data, control_data]).reset_index(drop=True)

    # Shuffle the data (optional)
    df = df.sample(frac=1, random_state=42).sort_values(by='timestamp').reset_index(drop=True)

    # Apply shifts and random noise to the units
    for iunit, shift in zip(df.unit.unique(), shifts):
        n_rows = df.loc[df.unit == iunit, 'metric'].shape[0]
        df.loc[df.unit == iunit, 'metric'] += shift + np.random.normal(0, random_noise_scale, n_rows)

    return df

# Example usage
if __name__ == "__main__":
    start_date_pre = "2023-10-01"
    end_date_pre = "2023-10-31"
    start_date_post = "2023-11-01"
    end_date_post = "2023-11-30"
    treatment_unit = "Treated_Unit"
    control_units = ["Control_Unit_1", "Control_Unit_2", "Control_Unit_3"]
    metric_slope = 0.5
    treatment_shift = 20
    random_noise_scale = 3
    shifts = [30, 20, 10, 0]

    df = generate_trend_dataset(start_date_pre, end_date_pre, start_date_post, end_date_post, treatment_unit, control_units,
                         metric_slope, treatment_shift, random_noise_scale, shifts)

    print(df.head())

    
def generate_dataset_sine(start_date_pre, end_date_pre, start_date_post, end_date_post, treatment_unit,
                          control_units, treatment_shift, random_noise_scale, shifts):
    """
    Generate a synthetic dataset for unit testing using a sine wave for the metric values.

    Parameters:
    - start_date_pre: str, the start date of the pre-treatment period.
    - end_date_pre: str, the end date of the pre-treatment period.
    - start_date_post: str, the start date of the post-treatment period.
    - end_date_post: str, the end date of the post-treatment period.
    - treatment_unit: str, the name of the treatment unit.
    - control_units: list of str, the names of the control units.
    - treatment_shift: float, the rigid shift post-treatment.
    - random_noise_scale: float, the scale of the random noise.
    - shifts: list of float, the shifts to apply to each unit.

    Returns:
    - pd.DataFrame, the synthetic dataset.
    """
    # Generate pre-treatment and post-treatment timestamps
    pre_period = pd.date_range(start=start_date_pre, end=end_date_pre)
    post_period = pd.date_range(start=start_date_post, end=end_date_post)

    def generate_unit_data(unit_name, is_treated=False):
        """
        Generate synthetic data for a single unit using a sine wave for the metric values.

        Parameters:
        - unit_name: str, the name of the unit.
        - is_treated: bool, whether the unit is treated.

        Returns:
        - pd.DataFrame, the synthetic data for the unit.
        """
        # Pre-treatment period
        pre_metrics = np.sin(np.linspace(0, 2 * np.pi * len(pre_period) / 7, len(pre_period)))
        # Post-treatment period
        if is_treated:
            post_metrics = np.sin(np.linspace(0, 2 * np.pi * len(post_period) / 7, len(post_period))) + treatment_shift
        else:
            post_metrics = np.sin(np.linspace(0, 2 * np.pi * len(post_period) / 7, len(post_period)))

        # Combine pre- and post-periods
        metrics = np.concatenate([pre_metrics, post_metrics])
        timestamps = np.concatenate([pre_period, post_period])

        # Create DataFrame
        return pd.DataFrame({
            "timestamp": timestamps,
            "unit": unit_name,
            "metric": metrics
        })

    # Generate data for the treatment unit
    treated_data = generate_unit_data(treatment_unit, is_treated=True)

    # Generate data for the control units
    control_data = pd.concat([generate_unit_data(unit) for unit in control_units])

    # Combine all data
    df = pd.concat([treated_data, control_data]).reset_index(drop=True)

    # Shuffle the data (optional)
    df = df.sample(frac=1, random_state=42).sort_values(by='timestamp').reset_index(drop=True)

    # Apply shifts and random noise to the units
    for iunit, shift in zip(df.unit.unique(), shifts):
        n_rows = df.loc[df.unit == iunit, 'metric'].shape[0]
        df.loc[df.unit == iunit, 'metric'] += shift + np.random.normal(0, random_noise_scale, n_rows)

    return df

# Example usage
if __name__ == "__main__":
    start_date_pre = "2023-10-01"
    end_date_pre = "2023-10-31"
    start_date_post = "2023-11-01"
    end_date_post = "2023-11-30"
    treatment_unit = "Treated_Unit"
    control_units = ["Control_Unit_1", "Control_Unit_2", "Control_Unit_3"]
    treatment_shift = 20
    random_noise_scale = 3
    shifts = [30, 20, 10, 0]

    df = generate_dataset_sine(start_date_pre, end_date_pre, start_date_post, end_date_post, treatment_unit, control_units,
                               treatment_shift, random_noise_scale, shifts)

    print(df.head())
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import nnls

def create_synthetic_control(df, time_col, unit_col, metric_col, treatment_unit, treatment_timestamp):
    """
    Create a synthetic control unit by finding the weights for the control units that best match the pre-treatment outcomes of the treated unit.

    Parameters:
    - df: pandas DataFrame containing the data.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - metric_col: str, the name of the metric column to be analyzed.
    - treatment_unit: str, the unit that received the treatment.
    - treatment_timestamp: Timestamp, the first timestamp of the treatment period.

    Returns:
    - synthetic_control: pandas DataFrame, the synthetic control unit.
    - weights: numpy array, the weights for the control units.
    """
    # Ensure the time column is in datetime format
    df[time_col] = pd.to_datetime(df[time_col])

    # Filter the data for the pre-treatment period
    pre_treatment_df = df[df[time_col] < treatment_timestamp]

    # Get the treated unit data
    treated_unit_data = pre_treatment_df[pre_treatment_df[unit_col] == treatment_unit]

    # Get the control units data
    control_units = pre_treatment_df[pre_treatment_df[unit_col] != treatment_unit]
    control_units_pivot = control_units.pivot(index=time_col, columns=unit_col, values=metric_col)

    # Ensure the control units pivot table has the same time points as the treated unit
    control_units_pivot = control_units_pivot.reindex(treated_unit_data[time_col])

    # Get the treated unit outcomes
    treated_unit_outcomes = treated_unit_data.set_index(time_col)[metric_col]

    #print(f"{treated_unit_data.shape=}") # DEBUG CARLO
    #print(f"{control_units.shape=}") # DEBUG CARLO
    #print(f"{control_units_pivot.shape=}") # DEBUG CARLO
    #display(control_units_pivot.head(2)) # DEBUG CARLO
    #print(f"{treated_unit_outcomes.shape=}") # DEBUG CARLO
    #display(treated_unit_outcomes.head(2)) # DEBUG CARLO


    # Solve for the weights using non-negative least squares
    X = control_units_pivot.values#.T
    y = treated_unit_outcomes.values
    #print(f"{X.shape=}") # DEBUG CARLO
    #print(f"{y.shape=}") # DEBUG CARLO
    weights, _ = nnls(X, y)

    # Create the synthetic control unit
    synthetic_control = control_units_pivot @ weights
    synthetic_control = pd.DataFrame({time_col: control_units_pivot.index, metric_col: synthetic_control})

    return synthetic_control, weights


def estimate_synthetic_did(df, time_col, unit_col, metric_col, treatment_unit, treatment_timestamp, clustered_errors=False, model_summary=False):
    """
    Estimate the synthetic DID effect by comparing the post-treatment outcomes of the treated unit to the synthetic control unit.

    Parameters:
    - df: pandas DataFrame containing the data.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - metric_col: str, the name of the metric column to be analyzed.
    - treatment_unit: str, the unit that received the treatment.
    - treatment_timestamp: Timestamp, the first timestamp of the treatment period.
    - clustered_errors: boolean, whether to use clustered standard errors.
    - model_summary: boolean, whether to print the model summary.

    Returns:
    - did_coefficient: float, the coefficient of the DID estimate.
    - p_value: float, the p-value of the DID estimate.
    - confidence_interval: tuple, the confidence interval of the DID estimate.
    """
    # Ensure the time column is in datetime format
    df[time_col] = pd.to_datetime(df[time_col])

    # Create the synthetic control unit
    synthetic_control, _ = create_synthetic_control(df, time_col, unit_col, metric_col, treatment_unit, treatment_timestamp)

    # Combine the synthetic control unit with the original data
    synthetic_control['unit'] = 'Synthetic_Control'
    combined_df = pd.concat([df, synthetic_control], ignore_index=True)

    # Prepare the data for fixed effects regression
    prepared_df = prepare_fixed_effects_data(combined_df, time_col, unit_col, metric_col, treatment_timestamp, treatment_unit)

    # Fit the fixed effects regression model
    did_coefficient, p_value, confidence_interval, model = fit_fixed_effects_model(prepared_df, metric_col, time_col, clustered_errors, model_summary)

    return did_coefficient, p_value, confidence_interval
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import nnls

def placebo_test_sdid(df, time_col, unit_col, metric_col, treatment_unit, treatment_timestamp, 
                      num_placebos=10, clustered_errors=True, verbose=False):
    """
    Perform placebo tests for Synthetic Difference-in-Differences (SDID) by applying the treatment 
    to untreated units at different points in time.

    Parameters:
    - df: pandas DataFrame containing the data.
    - time_col: str, the name of the time column.
    - unit_col: str, the name of the unit column.
    - metric_col: str, the name of the metric column to be analyzed.
    - treatment_unit: str, the unit that received the treatment.
    - treatment_timestamp: Timestamp, the first timestamp of the treatment period.
    - num_placebos: int, number of placebo tests to conduct.
    - clustered_errors: boolean, whether to use clustered standard errors.
    - verbose: boolean, if True, print intermediate results.

    Returns:
    - placebo_results: DataFrame with placebo test results including p-values for each test.
    """
    
    # Ensure the time column is in datetime format
    df[time_col] = pd.to_datetime(df[time_col])
    
    # Get control units
    control_units = df[df[unit_col] != treatment_unit][unit_col].unique()

    # Store results
    results = []
    
    for _ in range(num_placebos):
        # Randomly select a placebo unit and placebo treatment time
        placebo_unit = np.random.choice(control_units)
        placebo_timestamp = np.random.choice(df[time_col].unique()[df[time_col] < treatment_timestamp])
        
        if verbose:
            print(f"Placebo test for unit {placebo_unit} at timestamp {placebo_timestamp}")
        
        # Estimate SDID using placebo unit and time
        _, p_value, _ = estimate_synthetic_did(
            df, 
            time_col, 
            unit_col, 
            metric_col, 
            treatment_unit=placebo_unit, 
            treatment_timestamp=placebo_timestamp, 
            clustered_errors=clustered_errors
        )
        
        results.append({
            'placebo_unit': placebo_unit,
            'placebo_timestamp': placebo_timestamp,
            'p_value': p_value
        })
    
    placebo_results = pd.DataFrame(results)
    
    # Calculate the proportion of significant placebo tests
    significance_level = 0.05  # Common choice for significance level
    significant_placebos = (placebo_results['p_value'] < significance_level).sum()
    proportion_significant = significant_placebos / num_placebos
    
    if verbose:
        print(f"Proportion of significant placebos at {significance_level} level: {proportion_significant:.2f}")
    
    return placebo_results

# Assuming estimate_synthetic_did function is defined as in your previous code
```
